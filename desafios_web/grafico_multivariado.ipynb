```python
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f (x , y):
  result = np.sin(x) + 2*np.cos(y)
  return result

# Armo los vectores que usaré de inputs en mi función multivarial

res = 100
x = np.linspace(-5 , 5 , num=res)
y = np.linspace(-5 , 5 , num=res)

x , y = np.meshgrid(x , y) # Mediante al función meshgrid unificamos cada punto del vector x con cada punto del vector y para luego mandarlos como argumentos a la función f

# Uso la función f para encontrar los valores de z
z = f(x , y)

# Le indicamos a matplotlib que vamos a graficar una superficie 3d. fig lo uso para la figura y ax para nuestros ejes

fig , ax = plt.subplots(subplot_kw={"projection":"3d"})

# defino la superficie en nuestro eje

surf = ax.plot_surface(x , y , z, cmap = cm.cool) # Con cmap elijo el mapa de color que quiero. En este caso lo traje de cm en matplotlib
fig.colorbar(surf) # Le agrego el detalle de la barra de color a mi grafica
```
